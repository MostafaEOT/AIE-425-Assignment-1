```python
import pandas as pd
import numpy as np

# Load the movie titles from your CSV file
file_path = '/content/movies_2023_2023.csv'  # Replace with the actual file path
movies_df = pd.read_csv(file_path)

# Ensure we have unique movie titles if duplicates are present
movies_df = movies_df.drop_duplicates(subset='primaryTitle').reset_index(drop=True)

# Extract the movie titles
movie_titles = movies_df['primaryTitle'].tolist()
num_movies = len(movie_titles)  # Number of unique movies

# Set the number of users
num_users = 100
user_ids = [f"User{i+1}" for i in range(num_users)]

# Generate random ratings (from 1 to 5) and NaNs to simulate missing ratings
user_ratings = np.random.choice([1, 2, 3, 4, 5, np.nan], size=(num_movies, num_users), p=[0.15, 0.15, 0.2, 0.2, 0.2, 0.1])

# Create a DataFrame with movie titles as the index and user ratings as columns
user_item_df = pd.DataFrame(user_ratings, columns=user_ids, index=movie_titles)

# Display a sample of the user-item matrix
print(user_item_df.head())

# Save the user-item matrix to a CSV file
user_item_df.to_csv("user_item_matrix.csv")
print("User-item matrix saved as 'user_item_matrix.csv'")
```

**Prediction**

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the user-item matrix (assuming it's saved in 'user_item_matrix.csv')
user_item_df = pd.read_csv("user_item_matrix.csv", index_col=0)

# Compute cosine similarity and Pearson correlation (as you did earlier)
# Cosine similarity calculation
user_ratings = user_item_df.transpose()  # Users as rows, movies as columns
user_ratings_filled = user_ratings.fillna(0)  # Fill NaN with 0 for similarity calculation
cosine_sim = cosine_similarity(user_ratings_filled)
user_similarity_df = pd.DataFrame(cosine_sim, index=user_ratings.index, columns=user_ratings.index)

# Pearson correlation calculation
user_correlation = user_item_df.corr(method='pearson')
user_correlation_df = user_correlation

# Function to predict ratings based on user similarity
def predict_ratings(user_item_df, similarity_df, user_id, similarity_type='cosine'):
    # Get the ratings of the user
    user_ratings = user_item_df[user_id]

    # Items that the user hasn't rated yet
    unrated_items = user_ratings[user_ratings.isnull()].index

    predictions = {}

    for item in unrated_items:
        # Get the users who have rated the item
        rated_users = user_item_df.loc[item].dropna().index

        # If there are no rated users for this item, skip to the next item
        if len(rated_users) == 0:
            continue

        # Get the similarity scores for the user with the users who rated the item
        similarity_scores = similarity_df[user_id][rated_users]

        # Filter out the NaN similarities (in case there are any)
        similarity_scores = similarity_scores.dropna()

        # Get the ratings given by these users for the item
        ratings = user_item_df.loc[item, rated_users]

        # Calculate the weighted average prediction
        if len(similarity_scores) > 0:
            weighted_ratings = np.dot(similarity_scores, ratings[rated_users].dropna())
            prediction = weighted_ratings / similarity_scores.sum()  # Normalize by the sum of the similarities
            predictions[item] = prediction

    return predictions

# Predict ratings for all users
all_user_predictions = {}

for user_id in user_item_df.columns:  # Loop through each user
    cosine_predictions = predict_ratings(user_item_df, user_similarity_df, user_id, similarity_type='cosine')
    pearson_predictions = predict_ratings(user_item_df, user_correlation_df, user_id, similarity_type='pearson')

    all_user_predictions[user_id] = {
        'cosine': cosine_predictions,
        'pearson': pearson_predictions
    }

# Display predictions for all users
for user_id, predictions in all_user_predictions.items():
    print(f"\nPredictions for {user_id} (Cosine Similarity):")
    print(predictions['cosine'])

    print(f"\nPredictions for {user_id} (Pearson Correlation):")
    print(predictions['pearson'])
```

**Prediction**

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the user-item matrix
user_item_df = pd.read_csv("user_item_matrix.csv", index_col=0)

# Fill NaN values with 0 for similarity calculations
user_item_df_filled = user_item_df.fillna(0)

# Calculate item-based cosine similarity
cosine_sim_matrix = cosine_similarity(user_item_df_filled)
cosine_sim_df = pd.DataFrame(cosine_sim_matrix, index=user_item_df.index, columns=user_item_df.index)

# Calculate item-based Pearson correlation
pearson_corr_df = user_item_df.transpose().corr(method='pearson')

# Function to predict a rating for a specific user and item based on similarity matrix
def predict_rating(user_id, item, similarity_df, user_item_df):
    # Get similarity scores for the target item
    item_similarities = similarity_df[item]

    # Get user's ratings for all items, excluding the target item
    user_ratings = user_item_df[user_id]

    # Identify items that the user has rated
    rated_items = user_ratings.dropna().index

    # Filter similarities to only include items the user has rated
    relevant_similarities = item_similarities[rated_items]

    # Filter user ratings to include only rated items
    relevant_ratings = user_ratings[rated_items]

    # Calculate the weighted sum of ratings
    weighted_sum = np.dot(relevant_similarities, relevant_ratings)

    # Sum of absolute similarity values for normalization
    sum_of_weights = relevant_similarities.abs().sum()

    # Avoid division by zero by returning NaN if sum_of_weights is zero
    if sum_of_weights == 0:
        return np.nan

    # Prediction: weighted average
    return weighted_sum / sum_of_weights

# Predict ratings for all users and items
all_predictions_cosine = {}
all_predictions_pearson = {}

for user_id in user_item_df.columns:
    user_predictions_cosine = {}
    user_predictions_pearson = {}

    # Loop through each item (movie) that the user has not rated
    for item in user_item_df.index:
        if pd.isna(user_item_df.loc[item, user_id]):  # Check if the rating is missing (NaN)
            # Predict rating based on cosine similarity
            cosine_prediction = predict_rating(user_id, item, cosine_sim_df, user_item_df)
            user_predictions_cosine[item] = cosine_prediction

            # Predict rating based on Pearson correlation
            pearson_prediction = predict_rating(user_id, item, pearson_corr_df, user_item_df)
            user_predictions_pearson[item] = pearson_prediction

    all_predictions_cosine[user_id] = user_predictions_cosine
    all_predictions_pearson[user_id] = user_predictions_pearson

# Display sample predictions for all users
for user_id, predictions in all_predictions_cosine.items():
    print(f"\nPredicted ratings for {user_id} using cosine similarity:")
    print(predictions)

for user_id, predictions in all_predictions_pearson.items():
    print(f"\nPredicted ratings for {user_id} using Pearson correlation:")
    print(predictions)
```
